```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
os.listdir()
```

```python
f = open("./test_out.txt", "r")

lines = [line.strip('\n').split(',') for line in f.readlines() if line.strip()]

raw_accel_data = [line[0:2] for line in lines]

raw_gyro_data = [[0,0]]
for line in lines:
    raw_gyro_data.append([float(line[4])+raw_gyro_data[-1][0], float(line[5])+raw_gyro_data[-1][1]])

filter_data = [line[2:4] for line in lines]

time_list = [0.0]
for line in lines:
    time_list.append(float(line[6])+time_list[-1])


raw_accel_data_mat = np.matrix(raw_accel_data)
filter_data_mat = np.matrix(filter_data).astype(float)
gyro_data_mat = np.matrix(raw_gyro_data)
```

```python
plt.subplots()
plt.rcParams['figure.figsize'] = [40,20]
plt.plot(time_list[1:], raw_accel_data_mat[:,0])
plt.plot(time_list[1:], filter_data_mat[:,0], lw=5)
plt.plot(time_list, gyro_data_mat[:,0], lw=5)

plt.title('roll angle', fontsize=32)
plt.legend(['accelerometer','filter','gyro'], prop={'size':32})
```

```python
plt.subplots()
plt.rcParams['figure.figsize'] = [40,20]
plt.plot(time_list[1:],raw_accel_data_mat[:,1])
plt.plot(time_list[1:],filter_data_mat[:,1], lw=5)
plt.plot(time_list, gyro_data_mat[:,1], lw=5)

plt.title('pitch angle', fontsize=32)
plt.legend(['accelerometer','filter','gyro'], prop={'size':32})
```
